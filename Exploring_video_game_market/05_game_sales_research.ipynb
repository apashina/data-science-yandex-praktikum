# Исследование продаж компьютерных игр

### Описание проекта

Интернет-магазин «Стримчик» продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation).

**Цель проекта** - выявить определяющие успешность игры закономерности, что позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

**Задачи:** 
- изучить общую информацию о предоставленных для анализа данных,
- выполнить предобработку данных,
- провести исследовательский анализ данных,
- составить портрет пользователя каждого региона,
- проверить гипотезы,
- сформулировать общий вывод.

**Описание данных**

Перед нами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и мы планируем кампанию на 2017-й. Данные за 2016 год могут быть неполными.

* `Name` — название игры,

* `Platform` — платформа,
* `Year_of_Release` — год выпуска,
* `Genre` — жанр игры,
* `NA_sales` — продажи в Северной Америке (миллионы проданных копий),
* `EU_sales` — продажи в Европе (миллионы проданных копий),
* `JP_sales` — продажи в Японии (миллионы проданных копий),
* `Other_sales` — продажи в других странах (миллионы проданных копий),
* `Critic_Score` — оценка критиков (максимум 100),
* `User_Score` — оценка пользователей (максимум 10),
* `Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

```python
# удаление строк с пропусками в столбце 'name'
data.dropna(subset = ['name'], inplace = True)
```

Необходимо контролировать долю удаленных наблюдений из исходного датафрейма, выполняя предобработку данных. 

```python
def lost_observations(df):
    """
    Функция получает на вход обработанный датасет. 
    Выводит на экран процент удаленных записей из исходного датасета.
    
    """    
    print(f'{(16715 - df.shape[0]) / 16715:.2%}')
```

```python
lost_observations(data)
```

В колонке `user_score` самыми популярными являются строковые значения в виде аббревиатуры 'tbd' - to be determined (будет определено), поэтому заменим их на NaN. Затем преобразуем тип данных в этой колонке в вещественный.

```python
# замена значений 'tbd' на NaN
data['user_score'].replace('tbd', np.nan, inplace = True)

# преобразование типа данных в вещественный
data['user_score'] = data['user_score'].astype('float64')
```

Пропущенные значения в этих колонках не будут обработаны, поскольку их слишком много, и заполнять их средним или медианным значениями, группируя по году выпуска или жанру, будет некорректно. Множество факторов влияет на субъективное мнение критиков и пользователей - год, платформа, рекламная кампания, жанр и др.

#### Влияние отзывов пользователей и критиков на продажи внутри одной популярной платформы

Игры, которые лучше продаются, имеют высокую оценку критиков. Чего нельзя сказать в случае с оценкой пользователей, т.к. есть игры с низким рейтингом, но продающиеся хорошо.

На продажи игр оказывают слабое влияние отзывы критиков, об этом говорит положительный коэффициент корреляции, равный 0.4. Связи между глобальными продажами и оценками пользователей не обнаружено.

Далее посмотрим на связь между отзывами и продажами игр на других платформах.

```python
# количество игр по платформам, которые оценены критиками
actual_data.groupby('platform')[['critic_score']].count().sort_values(by='critic_score')
```

```python
# количество игр по платформам, которые оценены пользователями
actual_data.groupby('platform')[['user_score']].count().sort_values(by='user_score')
```

```python
# построение диаграмм рассеяния и расчет коэффициентов корреляции между
# продажами и отзывами для платформ XOne, WiiU и 3DS
for platform in actual_data['platform'].unique():
    data_platform = actual_data.query('platform == @platform')
    if data_platform['user_score'].count() > 10:
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,3))
        sns.scatterplot(x='critic_score', y='total_sales', data=data_platform, ax=ax1)
        sns.scatterplot(x='user_score', y='total_sales', data=data_platform, ax=ax2)
        fig.suptitle(f'Диаграммы рассеяния суммарных продаж и отзывов для платформы {platform}', fontsize=12)

        critic_corr = data_platform['critic_score'].corr(data_platform['total_sales'])
        user_corr = data_platform['user_score'].corr(data_platform['total_sales'])
        print(f'Коэффициент корреляции между отзывами критиков и продажами на {platform}: {critic_corr:.2f}.')
        print(f'Коэффициент корреляции между отзывами пользователей и продажами на {platform}: {user_corr:.2f}.')
        print()
```

Связь отзывов об играх на платформе XOne с количеством продаж почти такая же, как и в случае платформы PS4. Продажи игр на платформе 3DS слабее коррелируют с оценкам критиков, но зато имеется слабая корреляция с оценками пользователей. Корреляция продаж игр на платформе WiiU с оценками пользователей незначительно превышает корреляцию с оценками критиков. Это отличие от остальных платформ, возможно, связано с тем, что для игр, выпущенных на платформе WiiU, оставлено меньше отзывов.

Таким образом, во всех рассмотренных случаях наблюдается слабая прямая связь между оценками критиков и продажами. А связь между оценками пользователей и продажами либо слабая, либо отсутствует. Можно предположить, что покупатели прислушиваются больше к критикам, чем к пользователям.

### Составление портрета пользователя каждого региона

Для пользователя каждого региона определим самые популярные платформы (топ-5) и опишем различия в долях продаж. Выделим самые популярные жанры (топ-5) и оценим влияние рейтинга ESRB на продажи в отдельном регионе.

#### Портрет пользователя для региона Северная Америка

```python
def user_portrait(region_sales):
    """
    Функция получает на вход название признака региона и возвращает 
    сводные таблицы и их визуализацию по пяти популярным платформам,
    жанрам и рейтингам ESRB.

    """
        
    fig, ax = plt.subplots(1, 3, figsize=(15, 3))
    region = {'na_sales': 'Северной Америке',
              'eu_sales': 'Европе',
              'jp_sales': 'Японии',
              'other_sales': 'других странах',
             }
    for index, ax in zip(['platform','genre','rating'], ax):
        region_pivot_table = (actual_data
                              .pivot_table(index=index, values=region_sales, aggfunc='sum')
                              .sort_values(by=region_sales, ascending=False)
                              .reset_index()
                              .head(5)
                             )
        all_pivot_table = (actual_data
                           .pivot_table(index=index, values='total_sales', aggfunc='sum')
                           .reset_index()
                          )
        for p in list(region_pivot_table[index]):
            criterion = region_pivot_table[index] == p
            criterion2 = all_pivot_table[index] == p
            region_pivot_table.loc[criterion, 'share_of_sales'] = round(float(region_pivot_table.loc[criterion][region_sales])
                                                                        / float(all_pivot_table.loc[criterion2]['total_sales']), 2)
        display(region_pivot_table)
        sns.barplot(data=region_pivot_table, x=index, y=region_sales, ax=ax);
        fig.suptitle(f'Топ 5 популярных платформ, жанров и рейтингов ESRB в {region[region_sales]}', fontsize=12)
```

```python
# вывод и визуализация топ 5 популярных платформ, 
# жанров и рейтингов ESRB для региона Северная Америка
user_portrait('na_sales')
```

Таким образом, в Северной Америке за актуальный период:
* самые популярные платформы - с большим отрывом PS4, XOne, потом идут X360, 3DS и PS3, причем доля от мировых суммарных продаж для платформ X360 и XOne составляет примерно 59%, для 3DS - 26%, для остальных платформ - около 33%,
* самые популярные жанры игр - Shooter, далее с небольшим отставанием Action, Sports, Role-Playing и Misc, причем доля от мировых суммарных продаж для жанра Shooter составляет 46%, для Sports и Misc - около 41%, для Action - 36%, для Role-Playing - 33%,
* больше всего продаж игр с рейтингом "для взрослых" (43%), затем идут игры "для всех" (38%), далее "подросткам" (39%) и "для всех от 10 лет и старше" (46%). В скобках отражена доля от суммарных мировых продаж.

### Проверка гипотез

#### Гипотеза о средних пользовательских рейтингах платформ XOne и PC

Гипотеза: средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Сформулируем нулевую и альтернативную гипотезы:
* Н0: средние пользовательские рейтинги платформ Xbox One и PC одинаковые,
* Н1: средние пользовательские рейтинги платформ Xbox One и PC разные.

Для проверки гипотезы о равенстве средних двух генеральных совокупностей будем использовать t-тест с помощью метода `ttest_ind()` пакета `scipy.stats`.

```python
# формирование выборок с пользовательскими рейтингами платформ XOne и PC
sample_1 = actual_data.query('platform == "XOne"')['user_score'].dropna()
sample_2 = actual_data.query('platform == "PC"')['user_score'].dropna()
```

```python
# размер выборки, выборочное среднее и выборочное стандартное отклонение
display(sample_1.describe()[:3])
display(sample_2.describe()[:3])
```

Поскольку размер выборок разный, то считаем дисперсии выборок неравными.

```python
# критический уровень статистической значимости
alpha = .05 
# проверка гипотезы о равенстве средних двух независимых совокупностей
results = st.ttest_ind(sample_1, sample_2, equal_var = False)
print(f'p-значение: {results.pvalue:.3f}')
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Таким образом, разница между средними пользовательскими рейтингами платформ Xbox One и PC статичтически незначима, при уровне значимости 5%. Поэтому предположение о равенстве средних рейтингов считаем верным.

#### Гипотеза о средних пользовательских рейтингах жанров Action и Sports

Гипотеза: средние пользовательские рейтинги жанров Action и Sports разные.

Сформулируем нулевую и альтернативную гипотезы:
* Н0: средние пользовательские рейтинги жанров Action и Sports одинаковые,
* Н1: средние пользовательские рейтинги жанров Action и Sports разные.


```python
# формирование выборок с пользовательскими рейтингами жанров Action и Sports
sample_1 = actual_data.query('genre == "Action"')['user_score'].dropna()
sample_2 = actual_data.query('genre == "Sports"')['user_score'].dropna()
# размер выборки, выборочное среднее и выборочное стандартное отклонение
display(sample_1.describe()[:3])
display(sample_2.describe()[:3])
```

Поскольку размер выборок разный, то считаем дисперсии выборок неравными.

```python
# критический уровень статистической значимости
alpha = .05 
# проверка гипотезы о равенстве средних двух независимых совокупностей
results = st.ttest_ind(sample_1, sample_2, equal_var = False)
print(f'p-значение: {results.pvalue:.3f}')
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Таким образом, при условии, что нулевая гипотеза верна, вероятность получить наблюдаемый результат очень мала, при уровне значимости 5%. Поэтому предположение о том, что средние пользовательские рейтинги жанров Action и Sports разные, верно.
